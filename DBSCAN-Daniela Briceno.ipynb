# Algoritmo DBSCAN en un Jupyter Notebook 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn import datasets
```

#### Se cargan los datos

```python
import pandas as pd

# Cargar los datos
df = pd.read_csv('AllData1.csv')
print(df.head())
```

```python
# Visualización
import matplotlib.pyplot as plt
plt.scatter(df['temp_aire'], df['vel_viento'])
plt.xlabel('temp_aire')
plt.ylabel('vel_viento')
plt.title('Scatter plot de los datos')
plt.show()
```

```python
# Crear instancia de DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Ajustar DBSCAN a los datos y obtener los clusters asignados
clusters = dbscan.fit_predict(df[['temp_aire', 'vel_viento']])

# Visualizar los datos con colores por cluster
plt.scatter(df['temp_aire'], df['vel_viento'], c=clusters, cmap='viridis')
plt.xlabel('temp_aire')
plt.ylabel('vel_viento')
plt.title('Scatter plot de los datos con colores por cluster')
plt.colorbar(label='Cluster')
plt.show()
```

### Interpretación y Discusión:
#### Los clusters identificados por DBSCAN se muestran en la visualización. Cada punto de datos se asigna a un cluster 
#### La pertinencia del método DBSCAN depende de la distribución de los datos y la naturaleza del problema.
